```python
# FOR PROCESS THE VALID CONFIGURATIONS AND ADD IT TO A EXISTING EXCEL FILE

import os
import pandas as pd

# Function to read the text file
def read_text_file(file_path):
    with open(file_path, 'r') as file:
        return file.read().split('============================================')

# Function to process the content of the file
def process_content_first_version(configs):
    valid_data = []
    
    for config in configs:
        if "Configurazione valida" in config:
            lines = config.strip().split('\n')
            thread_size = None
            block_size = None
            total_threads = None
            threads_per_block = None
            total_elements = None
            times = []

            for line in lines:
                if "Configurazione valida" in line:
                    parts = line.split(': ')[1].split(')')[0].strip('()').split(', ')
                    thread_size = tuple(map(int, parts))

                elif "threadSize:" in line:
                    parts = line.split(': ')[1].strip('().').split(', ')
                    thread_size = tuple(map(int, parts))

                elif "blockSize:" in line:
                    parts = line.split(': ')[1].strip('().').split(', ')
                    block_size = tuple(map(int, parts))

                elif "Total threads:" in line:
                    total_threads = int(line.split(': ')[1].strip())

                elif "Total threads for each block:" in line:
                    threads_per_block = int(line.split(': ')[1].strip())

                elif "Total elements to calculate:" in line:
                    total_elements = int(line.split(': ')[1].strip())

                elif "Time elapsed:" in line:
                    time_elapsed = float(line.split(': ')[1].split()[0])
                    times.append(time_elapsed)

            if all([thread_size, block_size, total_threads, threads_per_block, total_elements, times]):
                mean_time = sum(times) / len(times)
                valid_data.append((thread_size, block_size, total_threads, threads_per_block, total_elements, mean_time))
    
    return valid_data

# Function to create the DataFrame
def create_dataframe(valid_data):
    columns = ['Thread Size', 'Block Size', 'Total Threads', 'Threads per Block', 'Total Elements', 'Mean Time (ms)']
    df = pd.DataFrame(valid_data, columns=columns)
    return df

# Main function to process the file and save to Excel
def main_first_version(file_path, output_path):
    configs = read_text_file(file_path)
    valid_data = process_content_first_version(configs)
    valid_df = create_dataframe(valid_data)
    
    if os.path.exists(output_path):
        with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
            valid_df.to_excel(writer, sheet_name='Valid Configurations', index=False)
    else:
        with pd.ExcelWriter(output_path) as writer:
            valid_df.to_excel(writer, sheet_name='Valid Configurations', index=False)
    
    print(f"Saved to: {output_path}")

# Define the folder path and file lists
folder_path = 'C:/Users/ACER/Desktop/UNIGE MAGISTRALE 1 ANNO/1 SEMESTRE/HIGH PERFORMANCE COMPUTING/ESAME 2/Results'

file_list = [
    "risultati_use_pitched_memory_512.txt",
    "risultati_use_pitched_memory_1000.txt",
    "risultati_use_pitched_memory_1024.txt"
]

# Create an array with input and output file paths
file_paths = [(os.path.join(folder_path, file_name), os.path.join(folder_path, file_name.replace('.txt', '.xlsx'))) for file_name in file_list]

# Process each file and save to Excel
for input_path, output_path in file_paths:
    print(f"Processing file: {input_path}")
    main_first_version(input_path, output_path)

print("Processing complete.")
```

```python
# Funzione per processare il contenuto del file
def process_content_first_version(lines):
    valid_data = []
    invalid_data = []
    
    current_config = None
    current_times = []
    is_valid = False

    for line in lines:
        line = line.strip()
        print(f"Processing line: {line}")  # Debug: stampa la riga attuale
        if "Configurazione valida:" in line:
            if current_config and is_valid:
                mean_time = sum(current_times) / len(current_times)
                valid_data.append((current_config['thread_size'], 
                                    current_config['block_size'], 
                                    current_config['total_threads'], 
                                    current_config['threads_per_block'], 
                                    current_config['total_elements'], 
                                    mean_time))
                current_times = []
            is_valid = True
            parts = line.split(":")[1].split(").")[0].replace('(', '').replace(')', '').split(',')
            print(f"Valid configuration parts: {parts}")  # Debug: stampa le parti della configurazione valida
            if len(parts) == 2:
                thread_size = tuple(map(int, [p.strip() for p in parts]))
                current_config = {'thread_size': thread_size}
                print(f"Set current_config: {current_config}")  # Debug: stampa la configurazione attuale
        elif "Configurazione non valida:" in line:
            is_valid = False
            parts = line.split(":")[1].split(").")[0].replace('(', '').replace(')', '').split(',')
            print(f"Invalid configuration parts: {parts}")  # Debug: stampa le parti della configurazione non valida
            if len(parts) == 2:
                thread_size = tuple(map(int, [p.strip() for p in parts]))
                current_config = {'thread_size': thread_size}
                print(f"Set current_config: {current_config}")  # Debug: stampa la configurazione attuale
        elif "Motivo:" in line and not is_valid:
            try:
                blocks_exceeded = int(line.split('(')[1].split(')')[0])
                invalid_data.append((current_config['thread_size'], blocks_exceeded))
                print(f"Added invalid data: {invalid_data[-1]}")  # Debug: stampa i dati non validi aggiunti
            except (ValueError, IndexError):
                print(f"Error parsing line: {line}")
        elif "threadSize:" in line and is_valid:
            current_config['thread_size'] = tuple(map(int, line.split(':')[1].strip().strip('()').split(', ')))
            print(f"Updated thread_size in current_config: {current_config['thread_size']}")  # Debug: stampa la thread size aggiornata
        elif "blockSize:" in line and is_valid:
            current_config['block_size'] = tuple(map(int, line.split(':')[1].strip().strip('()').split(', ')))
            print(f"Updated block_size in current_config: {current_config['block_size']}")  # Debug: stampa la block size aggiornata
        elif "Total threads:" in line and is_valid:
            current_config['total_threads'] = int(line.split(':')[1].strip())
            print(f"Updated total_threads in current_config: {current_config['total_threads']}")  # Debug: stampa il numero totale di thread aggiornato
        elif "Total threads for each block:" in line and is_valid:
            current_config['threads_per_block'] = int(line.split(':')[1].strip())
            print(f"Updated threads_per_block in current_config: {current_config['threads_per_block']}")  # Debug: stampa i thread per blocco aggiornati
        elif "Total elements to calculate:" in line and is_valid:
            current_config['total_elements'] = int(line.split(':')[1].strip())
            print(f"Updated total_elements in current_config: {current_config['total_elements']}")  # Debug: stampa il numero totale di elementi aggiornato
        elif "Time elapsed:" in line and is_valid:
            time_elapsed = float(line.split(':')[1].strip().split()[0])
            current_times.append(time_elapsed)
            print(f"Added time_elapsed: {time_elapsed}")  # Debug: stampa il tempo aggiunto
    
    if current_config and is_valid:
        mean_time = sum(current_times) / len(current_times)
        valid_data.append((current_config['thread_size'], 
                            current_config['block_size'], 
                            current_config['total_threads'], 
                            current_config['threads_per_block'], 
                            current_config['total_elements'], 
                            mean_time))
        print(f"Added final valid data: {valid_data[-1]}")  # Debug: stampa i dati validi finali aggiunti

    return valid_data, invalid_data
```

```python
import os
import pandas as pd

# Function to read the text file
def read_text_file(file_path):
    with open(file_path, 'r') as file:
        return file.read().split('============================================')

# Function to process the content of the file
def process_content_tile_matrix(configs):
    valid_data = []
    
    for config in configs:
        if "Testing tile_size" in config:
            lines = config.strip().split('\n')
            tile_size = None
            matrix_size = None
            block_size = None
            times = []

            for line in lines:
                if "Testing tile_size=" in line:
                    parts = line.split(' ')[1:]
                    tile_size = int(parts[0].split('=')[1])
                    matrix_size = int(parts[2].split('=')[1])

                elif "blockSize:" in line:
                    parts = line.split(': ')[1].strip('().').split(', ')
                    block_size = tuple(map(int, parts))

                elif "Time elapsed:" in line:
                    time_elapsed = float(line.split(': ')[1].split()[0])
                    times.append(time_elapsed)

            if all([tile_size, matrix_size, block_size, times]):
                mean_time = sum(times) / len(times)
                valid_data.append((tile_size, matrix_size, block_size, mean_time))
    
    return valid_data

# Function to create the DataFrame
def create_dataframe(valid_data):
    columns = ['Tile Size', 'Matrix Size', 'Block Size', 'Mean Time (ms)']
    df = pd.DataFrame(valid_data, columns=columns)
    return df

# Main function to process the files and save to a single Excel sheet
def main_tile_matrix(file_paths, output_path):
    all_valid_data = []
    
    for file_path in file_paths:
        configs = read_text_file(file_path)
        valid_data = process_content_tile_matrix(configs)
        all_valid_data.extend(valid_data)
    
    valid_df = create_dataframe(all_valid_data)
    
    if os.path.exists(output_path):
        with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
            valid_df.to_excel(writer, sheet_name='Tile Matrix Configurations', index=False)
    else:
        with pd.ExcelWriter(output_path) as writer:
            valid_df.to_excel(writer, sheet_name='Tile Matrix Configurations', index=False)
    
    print(f"Saved to: {output_path}")

# Define the folder path and file lists
folder_path = 'C:/Users/ACER/Desktop/UNIGE MAGISTRALE 1 ANNO/1 SEMESTRE/HIGH PERFORMANCE COMPUTING/ESAME 2/Results'

file_list2 = [
    "risultati_use_shared_memory_2_2000.txt",
    "risultati_use_shared_memory_4_2000.txt",
    "risultati_use_shared_memory_8_2000.txt",
    "risultati_use_shared_memory_16_2000.txt",
    "risultati_use_shared_memory_16_3000.txt",
    "risultati_use_shared_memory_16_5000.txt",
    "risultati_use_shared_memory_32_2000.txt",
    "risultati_use_shared_memory_32_3000.txt",
    "risultati_use_shared_memory_32_5000.txt",
    "risultati_use_shared_unified_memory_32_2000.txt",
    "risultati_use_shared_unified_memory_32_3000.txt",
    "risultati_use_shared_unified_memory_32_5000.txt"
]

# Create an array with input file paths
file_paths2 = [os.path.join(folder_path, file_name) for file_name in file_list2]

# Define the output path
output_path = os.path.join(folder_path, 'tile_matrix_configurations.xlsx')

# Process the files and save to Excel
main_tile_matrix(file_paths2, output_path)

print("Processing complete.")
```

```python
import pandas as pd
import re

# Funzione per leggere i file di testo e calcolare il tempo medio di esecuzione
def parse_file(file_path, tile_size, matrix_size):
    with open(file_path, 'r') as file:
        content = file.read()
    
    # Trova tutti i tempi di esecuzione usando regex
    times = re.findall(r'Time elapsed: ([\d.]+) milliseconds', content)
    times = list(map(float, times))  # Converti i tempi in float
    mean_time = sum(times) / len(times)  # Calcola il tempo medio
    return {'Tile Size': tile_size, 'Matrix Size': matrix_size, 'Block Size': f'({tile_size}, {tile_size}, 1)', 'Mean Time (ms)': mean_time}

# Percorsi dei file di testo
file_paths = [
    ("C:\\Users\\ACER\\Desktop\\UNIGE MAGISTRALE 1 ANNO\\1 SEMESTRE\\HIGH PERFORMANCE COMPUTING\\ESAME 2\\Results\\extracted_data_txt\\risultati_use_shared_unified_memory_32_2000.txt", 32, 2000),
    ("C:\\Users\\ACER\\Desktop\\UNIGE MAGISTRALE 1 ANNO\\1 SEMESTRE\\HIGH PERFORMANCE COMPUTING\\ESAME 2\\Results\\extracted_data_txt\\risultati_use_shared_unified_memory_32_3000.txt", 32, 3000),
    ("C:\\Users\\ACER\\Desktop\\UNIGE MAGISTRALE 1 ANNO\\1 SEMESTRE\\HIGH PERFORMANCE COMPUTING\\ESAME 2\\Results\\extracted_data_txt\\risultati_use_shared_unified_memory_32_5000.txt", 32, 5000)
]

# Lista per memorizzare i risultati
results = []

# Leggi i file e calcola i tempi medi
for file_path, tile_size, matrix_size in file_paths:
    result = parse_file(file_path, tile_size, matrix_size)
    results.append(result)

# Crea un DataFrame
df = pd.DataFrame(results)

# Salva il DataFrame in un file Excel
output_file = "C:\\Users\\ACER\\Desktop\\UNIGE MAGISTRALE 1 ANNO\\1 SEMESTRE\\HIGH PERFORMANCE COMPUTING\\ESAME 2\\Results\\extracted_data_table\\shared_unified_memory_results.xlsx"
df.to_excel(output_file, index=False)

# Visualizza il DataFrame
print(df)
```
